```python
def check_and_create_database_folder():
    """
    Checks if a directory named 'DATABASE' exists in the same directory as the current Jupyter Notebook.
    If the directory does not exist, it creates the directory.
    """
    # Get the directory of the current Jupyter Notebook
    notebook_dir = os.path.dirname(os.path.abspath("__file__"))
    
    # Define the path to the 'DATABASE' folder
    database_folder_path = os.path.join(notebook_dir, 'DATABASE')
    
    # Check if the 'DATABASE' folder exists
    if not os.path.isdir(database_folder_path):
        # Create the 'DATABASE' folder
        os.makedirs(database_folder_path)
        print(f"'DATABASE' folder created at: {database_folder_path}")
    else:
        print(f"'DATABASE' folder already exists at: {database_folder_path}")

# Run the function
check_and_create_database_folder()
```

```python
from PyPDF2 import PdfReader
```

```python
def pdfs_to_text(directory, output_file):
    """
    Converts all PDFs in a specified directory to a single text file.
    
    Parameters:
    - directory (str): Path to the directory containing PDF files.
    - output_file (str): Path to the output text file.
    """
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for filename in os.listdir(directory):
            if filename.endswith(".pdf"):
                pdf_path = os.path.join(directory, filename)
                reader = PdfReader(pdf_path)
                for page in reader.pages:
                    text = page.extract_text()
                    if text:
                        outfile.write(text + "\n")
    print(f"All PDFs have been successfully combined into {output_file}")

# pdfs_to_text('DATABASE/Poems', 'combined_Elytis.txt')
```
